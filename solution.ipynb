processing the data for training on randomforestclassifier

```python
results = df['Result']
ids = df['key']
df = df.drop(['Result'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df,results, test_size=0.2, train_size=0.8, random_state=32, shuffle=True)
```

```python
x_test.head()
```

RandomForestClassifier accuracy_score: 97%

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn import metrics
```

```python
rf = RandomForestClassifier(n_estimators=65)
rf.fit(X=x_train, y=y_train)
```

```python
y_pred = rf.predict(x_test)
print('recall score:',metrics.recall_score(y_test, y_pred, pos_label='positive', average='micro'))
print('precision score:',metrics.precision_score(y_test, y_pred, pos_label='positive', average='micro'))
print('f1 score:',metrics.f1_score(y_test, y_pred, pos_label='positive', average='micro'))
print(classification_report(y_test, y_pred))
print('confusion matrix:',confusion_matrix(y_test, y_pred))
```
